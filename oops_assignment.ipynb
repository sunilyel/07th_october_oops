```python
# 01) Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. 
        #Implement methods to deposit and withdraw money from the account

class bank_account:
    def __init__(self, name, acc_number, balance = 0):
        self.name = name
        self.acc_number = acc_number
        self.balance = balance
    
    def deposit_amount(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount} succesfully")
        else:
            print(f"Invalid amount for deposit.")
    
    def withdraw_amount(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawn {amount} successfully.")
        else:
            print("Insufficient funds or invalid amount for withdrawal.")
    
    def available_balance(self):
        print(f"Account Balance: {self.balance}")
        
account = bank_account("38825310", "pw skills", 1000)
account.available_balance()
account.deposit_amount(500)
account.available_balance()
account.withdraw_amount(200)
account.available_balance()
account.withdraw_amount(1500)
account.available_balance()
```

```python
# 02) Employee Management Create a class representing an employee with attributes like employee ID, name, and salary.
        #Implement methods to calculate the yearly bonus and display employee details

class employee_data:
    def __init__(self, emp_ID, name, salary):
        self.emp_ID = emp_ID
        self.name = name
        self.salary = salary
    
    def yearly_bonus(self, bonus_percentage):
        bonus = (bonus_percentage/100) * self.salary
        print(f"Total bonus is {bonus}")
        
    def emp_details(self):
        print(f"Employee name : {self.name}")
        print(f"Employee ID : {self.emp_ID}")
        print(f"Employee salary : {self.salary}")
    
employee1 = employee_data(24241, "pwskills", 20000)
employee2 = employee_data(25251, "iNeuron", 20001)

employee1.yearly_bonus(10)
employee1.emp_details()
print()
employee2.yearly_bonus(11)
employee2.emp_details()
```

```python
# 03) Vehicle Rental Create a class representing a vehicle rental system.
        #Implement methods to rent a vehicle, return a vehicle, and display available vehicles

class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles = {"001": "Car", "002": "Bike", "003": "Scooter"}

    def rent_vehicle(self, vehicle_id, vehicle_type):
        if vehicle_id in self.available_vehicles and self.available_vehicles[vehicle_id] == vehicle_type:
            del self.available_vehicles[vehicle_id]
            print(f"Vehicle {vehicle_id} rented successfully.")
        else:
            print(f"Vehicle {vehicle_id} not available for rent.")

    def return_vehicle(self, vehicle_id, vehicle_type):
        if vehicle_id not in self.available_vehicles:
            self.available_vehicles[vehicle_id] = vehicle_type
            print(f"Vehicle {vehicle_id} returned successfully.")
        else:
            print(f"Vehicle {vehicle_id} is already available.")

    def display_available_vehicles(self):
        if self.available_vehicles:
            print("Available Vehicles:")
            for vehicle_id, vehicle_type in self.available_vehicles.items():
                print(f"ID: {vehicle_id}, Type: {vehicle_type}")
        else:
            print("No vehicles available for rent.")


rental_system = VehicleRentalSystem()
rental_system.display_available_vehicles()
rental_system.rent_vehicle("001", "Car")
print()
rental_system.display_available_vehicles()
rental_system.rent_vehicle("001", "Car")
print()
rental_system.display_available_vehicles()
rental_system.return_vehicle("001", "Car")
print()
rental_system.display_available_vehicles()
rental_system.return_vehicle("002", "Bike")
print()
rental_system.display_available_vehicles()
```

```python
# 04) Library Catalogue Create classes representing a library and a book.
        #Implement methods to add books to the library, borrow books, and display available books

class books:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.isAvailable = True
        
    def __str__(self):
        return f"Title : {self.title} | Author : {self.author}"
        
class library:
    def __init__(self):
        self.available_books = []
    
    def add_book(self, book):
        self.available_books.append(book)
        print(f"Book '{book.title}' by {book.author} added to the library.")
        
    def borrow_book(self, title):
        for book in self.available_books:
            if book.title == title and book.isAvailable:
                book.isAvailable = False
                print(f"Book '{title}' borrowed successfully.")
                return
        print(f"Book '{title}' is either not available or does not exist in the library.")
            
    def avail_books(self):
        available_book = [book for book in self.available_books if book.isAvailable]
        if available_book:
            print("Available books:")
            for book in available_book:
                print(book)
        else:
            print("No books available in the library.")
            
            
user1 = library()    

book1 = books("The Great Gatsby", "F. Scott Fitzgerald")
book2 = books("To Kill a Mockingbird", "Harper Lee")
book3 = books("1984", "George Orwell")

user1.add_book(book1)
user1.add_book(book2)
user1.add_book(book3)
print()
user1.avail_books()
print()
user1.borrow_book("1984")
print()
user1.avail_books()
```

```python
# 05) Product Inventory Create classes representing a product and an inventory system.
        #Implement methods to add products to the inventory, update product quantity, and display available products

class Product:
    def __init__(self, id, name, price, quantity):
        self.id = id
        self.name = name
        self.price = price
        self.quantity = quantity
    
    def __str__(self):
        return f"{self.id}: {self.name}, ${self.price}, Quantity: {self.quantity}"

class Inventory:
    def __init__(self):
        self.products = []
    
    def add_product(self, product):
        self.products.append(product)
    
    def update_quantity(self, product_id, new_quantity):
        for product in self.products:
            if product.id == product_id:
                product.quantity = new_quantity
                return True
        return False
    
    def display_products(self):
        print("Available Products:")
        for product in self.products:
            print(product)

inventory = Inventory()

product1 = Product(1, "Laptop", 1200, 10)
product2 = Product(2, "Phone", 800, 15)
inventory.add_product(product1)
inventory.add_product(product2)

inventory.display_products()
print()
inventory.update_quantity(1, 5)
inventory.display_products()
```

```python
# 06) Shape Calculation Create a class representing a shape with attributes like length, width, and height.
        #Implement methods to calculate the area and perimeter of the shape

class Shape:
    def __init__(self, length=0, width=0, height=0):
        self.length = length
        self.width = width
        self.height = height
    
    def calculate_area(self):
        raise NotImplementedError("Subclasses must implement calculate_area method.")
    
    def calculate_perimeter(self):
        raise NotImplementedError("Subclasses must implement calculate_perimeter method.")

class Rectangle(Shape):
    def calculate_area(self):
        return self.length * self.width
    
    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

class Triangle(Shape):
    def calculate_area(self):
        s = (self.length + self.width + self.height) / 2
        return (s * (s - self.length) * (s - self.width) * (s - self.height)) ** 0.5
    
    def calculate_perimeter(self):
        return self.length + self.width + self.height

rectangle = Rectangle(length=4, width=6)
print("Rectangle Area:", rectangle.calculate_area())
print("Rectangle Perimeter:", rectangle.calculate_perimeter())
print()
triangle = Triangle(length=3, width=4, height=5)
print("Triangle Area:", triangle.calculate_area())
print("Triangle Perimeter:", triangle.calculate_perimeter())
```

```python
# 07) Student Management Create a class representing a student with attributes like student ID, name, and grades.
        #Implement methods to calculate the average grade and display student details

class Student:
    def __init__(self, student_id, name, grades):
        self.student_id = student_id
        self.name = name
        self.grades = grades
    
    def calculate_average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)
    
    def display_details(self):
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        print(f"Grades: {self.grades}")
        print(f"Average Grade: {self.calculate_average_grade()}")

student1 = Student(1, "pw", [85, 90, 92, 88])
student1.display_details()
print()
student2 = Student(2, "skills", [75, 80, 85])
student2.display_details()
```

```python
# 09) Social Media Profile Create a class representing a social media profile with attributes like username and posts.
        #Implement methods to add posts, display posts, and search for posts by keyword

class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []
    
    def add_post(self, post_content):
        self.posts.append(post_content)
    
    def display_posts(self):
        if self.posts:
            print(f"Posts by {self.username}:")
            for index, post in enumerate(self.posts, start=1):
                print(f"Post {index}: {post}")
        else:
            print(f"No posts found for {self.username}")
    
    def search_posts_by_keyword(self, keyword):
        matching_posts = [post for post in self.posts if keyword.lower() in post.lower()]
        if matching_posts:
            print(f"Posts by {self.username} containing '{keyword}':")
            for index, post in enumerate(matching_posts, start=1):
                print(f"Post {index}: {post}")
        else:
            print(f"No posts found for {self.username} containing '{keyword}'")

profile1 = SocialMediaProfile("john_doe")
profile1.add_post("Hello everyone! This is my first post.")
profile1.add_post("I'm enjoying the weather today.")
profile1.display_posts()
print()
profile2 = SocialMediaProfile("jane_smith")
profile2.add_post("Excited to start my new job!")
profile2.add_post("Just had a delicious dinner with friends.")
profile2.display_posts()
print()
profile1.search_posts_by_keyword("weather")
print()
profile2.search_posts_by_keyword("job")
```

```python
# 10) To Do List Create a class representing a To Do list with attributes like tasks and due dates.
        #Implement methods to add tasks, mark tasks as completed, and display pending tasks

class ToDoList:
    def __init__(self):
        self.tasks = {}
    
    def add_task(self, task, due_date):
        self.tasks[task] = due_date
    
    def mark_task_completed(self, task):
        if task in self.tasks:
            del self.tasks[task]
        else:
            print(f"Task '{task}' not found in ToDo list.")
    
    def display_pending_tasks(self):
        if self.tasks:
            print("Pending Tasks:")
            for task, due_date in self.tasks.items():
                print(f"Task: {task}, Due Date: {due_date}")
        else:
            print("No pending tasks found.")

todo_list = ToDoList()
todo_list.add_task("Complete project proposal", "2024-02-20")
todo_list.add_task("Buy groceries", "2024-02-18")
todo_list.add_task("Call John", "2024-02-21")

todo_list.display_pending_tasks()
print()
todo_list.mark_task_completed("Buy groceries")
todo_list.display_pending_tasks()

todo_list.mark_task_completed("Send email to client")
```
